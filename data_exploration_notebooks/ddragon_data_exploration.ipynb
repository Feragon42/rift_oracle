Theres a URL to get the last version available for the dataset:

```python
import requests as re

last_versions = re.get('https://ddragon.leagueoflegends.com/realms/na.json').json()
last_patch = last_versions['n']['item']
print(last_patch)
```

```python
output_filename = f"ddragon_data_{last_patch}.tgz"
url = f"https://ddragon.leagueoflegends.com/cdn/dragontail-{last_patch}.tgz"
downloaded = False

##Download tgz
with re.get(url, stream=True) as r:
    r.raise_for_status()
    with open(output_filename, 'wb') as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)
    print(f"Downloaded {output_filename} successfully.")
    downloaded = True

##Extract tgz
if downloaded:
    import tarfile
    with tarfile.open(output_filename, 'r:gz') as tar:
        tar.extractall()
    print(f"Extracted {output_filename} successfully.")
```

Instead of download all the .tgz, I can download only the Champions and Items json. Even specific champions json.
As the DDRAGON version and the PATCH version didnt match in its naming, let's check one Champion skills in the last DDRAGON version (16.3.1) and compare it with the last PATCH version (26.13)

```python
import requests as re

# versions = re.get('https://ddragon.leagueoflegends.com/realms/na.json').json()
# last_version = versions['n']['item']

versions_list = re.get('https://ddragon.leagueoflegends.com/api/versions.json').json()
last_version = versions_list[0]
previous_version = versions_list[1]

champion_name = 'Brand'
champion_url_last = f'https://ddragon.leagueoflegends.com/cdn/{last_version}/data/en_US/champion/{champion_name}.json'
champion_data_last = re.get(champion_url_last).json()

champion_url_previous = f'https://ddragon.leagueoflegends.com/cdn/{previous_version}/data/en_US/champion/{champion_name}.json'
champion_data_previous = re.get(champion_url_previous).json()

print(f"Champion: {champion_name}")
#print(f"Stats: {champion_stats}")

print('In the patch, Brand E cost past from 70 / 75 / 80 / 85 / 90 to a fixed 90 mana cost:')

champion_spells_last = champion_data_last['data'][champion_name]['spells']
champion_spells_previous = champion_data_previous['data'][champion_name]['spells']

print(f"E (previous): {champion_spells_previous[2]}")
print(f"--Manacost (previous) {champion_spells_previous[2]['costBurn']}")
print(f"E (last): {champion_spells_last[2]}")
print(f"--Manacost (last) {champion_spells_last[2]['costBurn']}")

print('In the patch, Cassiopea Health per Level past from 104 to 98:')

champion_name = 'Cassiopeia'
champion_url_last = f'https://ddragon.leagueoflegends.com/cdn/{last_version}/data/en_US/champion/{champion_name}.json'
champion_data_last = re.get(champion_url_last).json() 
champion_url_previous = f'https://ddragon.leagueoflegends.com/cdn/{previous_version}/data/en_US/champion/{champion_name}.json'
champion_data_previous = re.get(champion_url_previous).json()

champion_stats_last = champion_data_last['data'][champion_name]['stats']
champion_stats_previous = champion_data_previous['data'][champion_name]['stats']

print(f"Cassiopeia (previous): {champion_stats_previous}")
print(f"--Health per Level (previous) {champion_stats_previous['hpperlevel']}")
print(f"Cassiopeia (last): {champion_stats_last}")
print(f"--Health per Level (last) {champion_stats_last['hpperlevel']}")
```
