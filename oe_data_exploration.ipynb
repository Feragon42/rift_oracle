First, let's check what we have in this csv

```python
import pandas as pd

ds = pd.read_csv("datasets/2023_LoL_esports_match_data_from_OraclesElixir.csv")
print('## Shape of the dataset:', ds.shape)
print('## Columns in the dataset:', len(ds.columns))
print('## First 5 rows of the dataset:\n', ds.head())

print('\n## Lista completa de columnas:')
for i, col in enumerate(ds.columns, start=1):
    print(f"{i:3d}. {col}")
```

With the head it looks the data is per player, I need to check one specific match to understand the data structure:

```python
match_info = ds[ds['gameid'] == 'ESPORTSTMNT06_2753012']
print(match_info[['gameid', 'league', 'datacompleteness', 'date', 'participantid', 'side', 'position', 'playername', 'teamname', 'champion']])
```

There are 12 rows per match. 10 for players and 2 for every team. I need to check the detail with one player of each team and the teams rows:

```python
pd.set_option('display.max_columns', None)
print(match_info[match_info['playername'].isin(['Wylenz', 'Anathar', float('nan')])].to_string(index=False))
```
